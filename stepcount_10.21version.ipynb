```python
import os

directory_list = []
for root, dirs, files in os.walk("/Users/yiran/Desktop/Step-Omics/step count data files", topdown=False):
    for name in dirs:
        directory_list.append(os.path.join(root, name))

import glob
import pandas as pd

step_count_activity=pd.DataFrame()
filenames=pd.DataFrame()
str='/*.xls'
paths=[x+str for x in directory_list]
paths.sort()
```

```python
for path in paths:
    filenames = glob.glob(path)
    for filename in filenames:
        df=pd.read_excel(filename, sheet_name='Activities')
        df['studyID']=path[54:58]
        step_count_activity=step_count_activity.append(df)
step_count_activity=step_count_activity.set_index('studyID')
step_count_activity
```

```python
step_count_activity.columns
```

```python
step_count_body=pd.DataFrame()
filenames=pd.DataFrame()
str='/*.xls'
paths=[x+str for x in directory_list]
paths.sort()
for path in paths:
    filenames = glob.glob(path)
    for filename in filenames:
        df=pd.read_excel(filename, sheet_name='Body')
        df['studyID']=path[54:58]
        step_count_body=step_count_body.append(df)
step_count_body=step_count_body.set_index('studyID')
step_count_body
```

```python
step_count_body.columns
```

```python
step_count_activityselect=step_count_activity[['Steps']].copy()
step_count_bodyselect=step_count_body[['Date','Weight','BMI']].copy()
step_count=pd.concat([step_count_activityselect, step_count_bodyselect], axis=1, sort=False)
step_count
```
